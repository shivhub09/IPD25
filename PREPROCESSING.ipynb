IMPORTING THE REQUIRED LIBRARIES

```python
from docx import Document
import os
import nltk
from nltk.corpus import stopwords
import string
```

IMPORTING THE DATASETS

```python
DATA_DIR = 'datasets'
```

```python
files = os.listdir(DATA_DIR)
```

GETTING THE CONTENT

```python
files = os.listdir(DATA_DIR)

contents = []

# Iterate through the files
for file_name in files:
    file_path = os.path.join(DATA_DIR, file_name)
    # Check if the item is a file
    if os.path.isfile(file_path) and file_name.endswith('.docx'):
        # Open the .docx file and read its contents
        doc = Document(file_path)


        for paragraph in doc.paragraphs:
                words = paragraph.text.lower().split(' ')
                contents.extend(words)
```

```python
def filter_words(words):
    # Get the list of English stopwords
    stop_words = set(stopwords.words('english'))
    
    # Remove punctuation and stopwords, and filter out numbers
    filtered_words = [word.strip(string.punctuation) for word in words if word.isalpha() and word not in stop_words]
    
    return filtered_words
```

```python
filtered_words = filter_words(contents)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Example documents
documents = ["This is the first document.",
             "This document is the second document.",
             "And this is the third one.",
             "Is this the first document?"]

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the documents
tfidf_matrix = vectorizer.fit_transform(documents)

# Get the feature names (terms)
feature_names = vectorizer.get_feature_names_out()

# Convert the TF-IDF matrix to a dense array for inspection
dense_matrix = tfidf_matrix.todense()

# Display the TF-IDF feature matrix
print("TF-IDF Feature Matrix:")
print(dense_matrix)
```
