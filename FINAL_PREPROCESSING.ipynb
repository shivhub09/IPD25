```python
from nltk.tokenize import word_tokenize
import re

def remove_numbers(input_list):
    tokenized_list = [word_tokenize(item) for item in input_list]

    cleaned_list = [' '.join(re.findall(r'\b\D+\b', ' '.join(tokens))) for tokens in tokenized_list]

    return cleaned_list

result_list = remove_numbers(tokens_lower)

print("List after removing numbers:", result_list)
```

```python
no_punctuations = [token for token in result_list if  token not in string.punctuation]
no_punctuations
```

```python
import re

def remove_symbols_from_list(input_list):
    pattern = re.compile(r'[\W_]+')

    cleaned_list = [re.sub(pattern, '', text) for text in input_list]

    return cleaned_list

result_list = remove_symbols_from_list(no_punctuations)

result_list
```

```python
def remove_duplicate_words(input_list):
    # Create an empty set to store unique words
    unique_words = set()

    # Create a new list to store the strings with duplicate words removed
    result_list = []

    for string in input_list:
        # Split each string into words
        words = string.split()

        # Remove duplicates while preserving the order
        unique_words_in_string = []
        for word in words:
            if word and word not in unique_words:
                unique_words.add(word)
                unique_words_in_string.append(word)

        # Join the unique words back into a string
        result_string = ' '.join(unique_words_in_string)

        # Append the result string to the new list
        result_list.append(result_string)

    return result_list

# Example usage:
result_list = remove_duplicate_words(tokens_no_stopwords)


result_list
```

```python
from nltk.stem import SnowballStemmer

def snowball_stemming_for_list(input_list):
    # Choose the language for the Snowball Stemmer
    language = 'english'
    
    # Create a Snowball Stemmer for the specified language
    stemmer = SnowballStemmer(language)
    
    # Create a new list to store the stemmed strings
    stemmed_list = []
    
    # Apply stemming to each string in the input list
    for text in input_list:
        # Tokenize the text into words and apply stemming
        stemmed_words = [stemmer.stem(word) for word in text.split()]
        
        # Join the stemmed words back into a string
        stemmed_text = ' '.join(stemmed_words)
        
        # Append the stemmed string to the new list
        stemmed_list.append(stemmed_text)
    
    return stemmed_list

# Example usage:
stemmed_list = snowball_stemming_for_list(result_list)


stemmed_list
```

```python
def remove_special_characters(input_string):
    # Use isalnum() to filter out non-alphanumeric characters
    cleaned_string = ''.join(char for char in input_string if char.isalnum())
    return cleaned_string

# Example usage:
original_string = "Hello @World!123"
cleaned_string = remove_special_characters(original_string)
print("Original String:", original_string)
print("Cleaned String:", cleaned_string)
```
