```python
from docx import Document
import os
from nltk.corpus import stopwords
import nltk

# Download NLTK stopwords (if not already downloaded)
nltk.download('stopwords')

def extract_text_from_docx(docx_path):
    text = ""
    doc = Document(docx_path)

    # Iterate through paragraphs and extract text
    for paragraph in doc.paragraphs:
        text += paragraph.text + '\n'

    return text

# Replace 'your_folder_path' with the path to the folder containing your Word files
folder_path = 'datasets'

corpus = []

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.docx'):
        docx_path = os.path.join(folder_path, filename)
        extracted_text = extract_text_from_docx(docx_path)
        
        # Append extracted text to the corpus list
        corpus.append(extracted_text)

# Concatenate the text from all documents into a single string
full_corpus_text = '\n'.join(corpus)

# Text Preprocessing - Remove unwanted and repetitive words (stopwords)
stop_words = set(stopwords.words('english'))
preprocessed_corpus = ' '.join([word for word in full_corpus_text.split() if word.lower() not in stop_words])

# Print the preprocessed corpus
print((preprocessed_corpus))
```
